A simple harmonic oscillator is given by the equation
$$x(t) = A \cos(\omega t + \phi)$$
where
$$\omega = \frac{2 \pi}{T}$$
$$A$$ is the amplitude,
$$T$$ is the period given in cycles per unit time,
and
$$\phi$$ is the phase.

The frequency $$f = 1 / T $$.
